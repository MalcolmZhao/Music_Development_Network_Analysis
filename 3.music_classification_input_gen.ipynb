```python
import numpy as np
import pandas as pd
import os
import pickle
```

```python
music_space = pd.read_csv("rearrange_data/music_influence_follow_feature.csv")
```

```python
music_space_transpos = music_space.pivot_table(index = "music_id", columns = "genre_id")
```

```python
music_feature = pd.read_csv("rearrange_data/music_features.csv")
```

```python
from itertools import permutations
genre_space_lst = list(permutations(range(4)))
genre_space_lst = [list(i) for i in genre_space_lst]
```

```python
music_id = music_feature["music_id"].tolist()
```

```python
feature_columns = ['danceability', 'energy', 'valence', 'tempo', 'loudness', 'mode', 'key',
       'acousticness', 'instrumentalness', 'liveness', 'speechiness',
       'explicit', 'duration_ms', 'popularity', 'musician_cnt']
```

```python
genre_dict = {
    "niche1": 0,
    "niche2": 1,
    "pop_rock": 2,
    "r_b": 3
}
```

```python
music_dense_feature = []
music_genre = np.zeros((len(music_id), 4))
music_influence_space = []
music_follow_space = []
cnt = 0
for idx, music in enumerate(music_id):
    cnt += 1
    music_piece = music_feature.loc[music_feature["music_id"] == music, :]
    target_feature = music_piece.loc[:, feature_columns].values[0]
    music_dense_feature.append(target_feature)
    
    target_space = music_space_transpos.loc[music, :]
    space_influence_tmp = []
    space_follow_tmp = []
    for space_combine in genre_space_lst:
        space_influence_tmp.append(target_space["music_influence_genre"][space_combine].values)
        space_follow_tmp.append(target_space["music_follow_genre"][space_combine].values)
    space_influence_tmp = np.array(space_influence_tmp)
    space_follow_tmp = np.array(space_follow_tmp)
    space_influence_tmp = space_influence_tmp.reshape((1, -1))
    space_follow_tmp = space_follow_tmp.reshape((1, -1))
    music_influence_space.append(space_influence_tmp)
    music_follow_space.append(space_follow_tmp)
    
    genre_val = genre_dict[music_piece["genre"].values[0]]
    music_genre[idx, genre_val] = 1
    if cnt % 100 == 0:
        print(cnt, music)
```

```python
music_dense_feature = np.array(music_dense_feature)
music_dense_feature.shape
```

```python
music_influence_space = np.array(music_influence_space)
music_influence_space.shape
```

```python
music_follow_space = np.array(music_follow_space)
music_follow_space.shape
```

```python
music_influence_space = music_influence_space.reshape((93878, 12, 8))
music_follow_space = music_follow_space.reshape((93878, 12, 8))
```

```python
idx = np.array(range(93878))
np.random.shuffle(idx)
```

```python
print(music_dense_feature.shape)
print(music_influence_space.shape)
print(music_follow_space.shape)
print(music_genre.shape)
```

```python
music_dense_feature = music_dense_feature[idx, :]
music_influence_space = music_influence_space[idx, :, :]
music_follow_space = music_follow_space[idx, :, :]
music_genre = music_genre[idx, :]
```

```python
music_classification_input = [music_dense_feature, music_influence_space, music_follow_space, music_genre]
with open(r'variables/music_classification_input.pkl', 'wb') as save_data:
    pickle.dump(music_classification_input, save_data) 
```
